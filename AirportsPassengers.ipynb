```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
def getCSV(csv):
    dfl = pd.read_csv(csv)
    return dfl
df = getCSV(r'C:\Users\Tom\OneDrive\Documents\APROJECTS\airportsPassengers.csv') 
print(df)
```

```python
filtered_df = df[df["Airport"]!="All main airports"]
filtered_df['Month'] = filtered_df['Month'].apply(lambda x: x[:8])
```

```python
#dub_df = filtered_df[filtered_df["Airport"]=="Dublin"]
cork_df = filtered_df[filtered_df["Airport"]=="Cork"]
shannon_df = filtered_df[filtered_df["Airport"]=="Shannon"]
knock_df = filtered_df[filtered_df["Airport"]=="Knock"]
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,6))

#dub_df.plot(x="Month", y="VALUE", marker='None', color='#267fb8', linestyle='-', linewidth=2.5, markersize=5, ax=ax, label="Dublin")
cork_df.plot(x="Month", y="VALUE", marker='None', color='#e50000', linestyle='-', linewidth=2.5, markersize=5, ax=ax, label="Cork")
shannon_df.plot(x="Month", y="VALUE", marker='None', color='#d2cb00', linestyle='-', linewidth=2.5, markersize=5, ax=ax, label="Shannon")
knock_df.plot(x="Month", y="VALUE", marker='None', color='#53ac52', linestyle='-', linewidth=2.5, markersize=5, ax=ax, label="Knock")


title_string = "Passengers handled by main airports"
ax.set_title(title_string)
ax.set_facecolor('#EEEEEE') 
#ax.set_ylim(110, 240)
plt.title(label=title_string, fontsize=14)
plt.xlabel('Month', fontsize=13)
plt.ylabel('Passengers (thousands)', fontsize=13)
plt.legend(title='', loc='upper left', fontsize='11')
plt.tight_layout()
plt.grid(True)

plt.subplots_adjust(left=0.1, right=0.90, top=0.90, bottom=0.15)
plt.text(0.7, -0.14, "Data Source: https://data.cso.ie/ \n\"TAM08 Passengers handled by main airports\"", ha='left', transform=ax.transAxes)

plt.savefig("Airports-Cork-Shannon-Knock.png")

plt.show()
```
