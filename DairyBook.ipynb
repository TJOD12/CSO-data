```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
def getCSV(csv):
    dfl = pd.read_csv(csv)
    return dfl
df = getCSV(r'C:\Users\Tom\OneDrive\Documents\APROJECTS\DairyProducts.csv') 
print(df)
```

```python
grouped = df.groupby('Product')
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,8))

colors=["#C5C80E","orange","blue"]
i=0
for group,data in grouped:
    plt.plot(data['Year'], data['VALUE'], label=group, color=colors[i], linewidth=2)
    i+=1

title_string = "Production of Dairy Products, Ireland, 1993 - 2021. (tonnes)"
plt.title(title_string)
plt.xlabel('Year')
plt.ylabel('Value (tonnes)')
plt.legend(title='Product', loc='upper left', fontsize='13')
#ax.set_ylim(20,75)
ax.set_facecolor('#EEEEEE')
plt.grid(True)

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.15)
plt.text(0.65, -0.11, "Data Source: https://data.cso.ie/ \n\"AKA03, Production of Dairy Products\"", ha='left', transform=ax.transAxes)

plt.savefig('Dairy-Products.png')

plt.show()
```
