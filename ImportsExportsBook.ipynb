```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
def getCSV(csv):
    dfl = pd.read_csv(csv)
    return dfl
    
csv = r'C:\Users\Tom\OneDrive\Documents\APROJECTS\imports-exports-services.csv'
df = getCSV(csv)
print(df)
```

```python
def getYears():
    distinct_years = set(df['Year'].tolist())
    all_years = [x for x in distinct_years]
    return all_years
years = getYears()
```

```python
imports_label = "Imports of Services"
imports_df = df[(df['Statistic Label'] == imports_label) & (df['Geographic Location'] == "All countries and international organisations")]
print(imports_df)
```

```python
fig, ax = plt.subplots(1,1, figsize = (12, 8))

imports_df.plot(x="Year", y="VALUE", kind="line", marker="None", linestyle="-", color="blue", ax=ax, label="Imports")

title_string = imports_label + ", 2003-2022."
plt.title(title_string)
plt.xlabel("Year")
plt.ylabel("Value (millions €)")
plt.xticks([int(year) for year in years])

ax.set_facecolor('#EEEEEE') 
#plt.grid(True)
plt.savefig("2003-2022-Total-Imports-Services.png")

plt.show()
```

```python
exports_label = "Exports of Services"
exports_df = df[(df['Statistic Label'] == exports_label) & (df['Geographic Location'] == "All countries and international organisations")]
print(exports_df)
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,8))

exports_df.plot(x="Year", y="VALUE", ax=ax, kind="line", marker="None", linestyle="-", color="red", label="Exports")
imports_df.plot(x="Year", y="VALUE", kind="line", marker="None", linestyle="-", color="blue", ax=ax, label="Imports")

title_string = "Imports and Exports of services, 2003-2022."
plt.title(title_string)
plt.xlabel("Year")
plt.ylabel("Value (millions €)")
plt.xticks([int(year) for year in years])

ax.set_facecolor('#EEEEEE') 
#plt.grid(True)
plt.savefig("2003-2022-Total-ImportsExports-Services.png")

plt.show()
```
