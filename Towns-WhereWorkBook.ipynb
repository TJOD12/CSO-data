```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
def getCSV(csv):
    dfl = pd.read_csv(csv)
    return dfl
    
df = getCSV(r'C:\Users\Tom\OneDrive\Documents\APROJECTS\F7149.20250511T110535.csv')
print(list(set(df['City of place of work'])))
```

```python
county = "Mayo"
work = "Galway city and suburbs, Co Galway"
filtered_df = df[df['Electoral Divisions'].str.contains(county)]
filtered_df['Electoral Divisions'] = filtered_df['Electoral Divisions'].apply(lambda x: x.split(", ")[0])
filtered_df = filtered_df[filtered_df['City of place of work']==work]
print(filtered_df)
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,25))

filtered_df.plot(x="Electoral Divisions", y="VALUE", kind="barh", width=0.75, edgecolor="black", color="#AA4455", ax=ax)

ax.set_facecolor('#EEEEEE')

plt.xlabel('Number of Workers', fontsize=14) 
plt.ylabel('Electoral Division', fontsize=14)
#plt.xlim(30000,70000)
plt.grid(axis='x') #plt.legend(loc='upper left', bbox_to_anchor=(.999, .96))
plt.legend().set_visible(False)
plt.xticks(rotation=0) 

plt.title("Population of "+county+" divisions who are working in Galway city and suburbs, 2022.")

plt.subplots_adjust(left=0.08, right=0.8, top=0.9, bottom=0.15) 
plt.text(.75, -0.045, "Data Source: https://data.cso.ie/ \nF7149 Population Usually Resident \nand Present in State", ha='left', transform=ax.transAxes)

plt.savefig('Town-Workers-'+county+"-"+work+'.png', bbox_inches='tight')

plt.show()
```
